## After performing feature selection in R. We are now ready to perform prediction. For this project linear regression and RandomForestRegressor is used to predict the bike usage.

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import preprocessing
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
import random
random.seed(123)
original = pd.read_csv("../Data/daily.csv")
```

## A function for spltting the dataset and performing the regression

```python
# read the train and test dataset
def function1(df1,model,test_prop=0.25):
    
    train=df1.sample(frac=1-test_prop)
    test=df1.drop(train.index)
    # shape of the dataset

    # seperate the independent and target variable on training data
    train_predictors_c=train.drop(columns=['casual','cnt','holiday','hum','temp','atemp'],axis=1)
    train_predictors_r=train.drop(columns=['registered','cnt','holiday','mnth','temp'],axis=1)
    train_target_c = train['casual']
    train_target_r = train['registered']

    # seperate the independent and target variable on testing data
    test_predictors_c=test.drop(columns=['casual','cnt','holiday','hum','temp','atemp'],axis=1)
    test_predictors_r=test.drop(columns=['registered','cnt','holiday','mnth','temp'],axis=1)
    test_target_c = test['casual']
    test_target_r = test['registered']

       
    min_max_scaler = preprocessing.MinMaxScaler()
    
    train_predictors_c = min_max_scaler.fit_transform(train_predictors_c.values)
    test_predictors_c = min_max_scaler.fit_transform(test_predictors_c.values)
    model=model.fit(train_predictors_c,train_target_c)
    predict_result_for_test_c = model.predict(test_predictors_c)
    
    train_predictors_r = min_max_scaler.fit_transform(train_predictors_r.values)
    test_predictors_r = min_max_scaler.fit_transform(test_predictors_r.values)
    model=model.fit(train_predictors_r,train_target_r)
    predict_result_for_test_r = model.predict(test_predictors_r)

    return test['cnt'],predict_result_for_test_r+predict_result_for_test_c
```

## Using Linear Regression

```python
gh=[]
gh1=[]
gh2=[]
or_c=original.drop(columns=["instant","dteday"],axis=1)
model = LinearRegression() 
for i in range(0,15):
    act_cnt1,pre_cnt1 = function1(or_c,model)
    act_cnt1=act_cnt1.values
    pre_cnt1=pre_cnt1
    f=[]
    for i in range(0,len(act_cnt1)):
        f.append((abs(act_cnt1[i]-pre_cnt1[i])*100)/act_cnt1[i])
    les10=0
    ab10=0
    ab20=0
    for i in range(0,len(f)):
        if f[i]<=10:
            les10+=1
        elif f[i] <= 20 and f[i] > 10 :
            ab10+=1
        else:
            ab20+=1
    gh.append(les10/len(f))
    gh1.append(ab10/len(f))
    gh2.append(ab20/len(f))
a=sum(gh)/15
b=sum(gh1)/15
c=sum(gh2)/15
print(a*100,"% of total prediction has error below10%. Where as",b*100,"% of total prediction has error range between 10% and 20%. Only",c*100,"% of total prediction has error above 20%")
```

```python
import matplotlib.pyplot as plt 
plt.plot(act_cnt1, label = "Actual")
plt.plot(pre_cnt1, label = "Predicted")
plt.xlabel('combined value of registred and casual bike usage')
plt.ylabel('Cnt')
plt.title('Combined value Prediction from Registered ans Casual bike usage ')
plt.legend()
plt.show()
```

## Using RandomForestRegressor

```python
g=[]
g1=[]
g2=[]
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
model = RandomForestRegressor(n_estimators = 500, random_state = 42) 
or_c=original.drop(columns=["instant","dteday"],axis=1)
for i in range(0,15):
    act_cnt2,pre_cnt2 = function1(or_c,model)
    act_cnt2=act_cnt2.values
    pre_cnt2=pre_cnt2
    f=[]
    for i in range(0,len(act_cnt2)):
        f.append((abs(act_cnt2[i]-pre_cnt2[i])*100)/act_cnt2[i])
    les10=0
    ab10=0
    ab20=0
    for i in range(0,len(f)):
        if f[i]<=10:
            les10+=1
        elif f[i] <= 20 and f[i] > 10 :
            ab10+=1
        else:
            ab20+=1
    g.append(les10/len(f))
    g1.append(ab10/len(f))
    g2.append(ab20/len(f))
a=sum(g)/15
b=sum(g1)/15
c=sum(g2)/15
print(a*100,"% of total prediction has error below10%. Where as",b*100,"% of total prediction has error range between 10% and 20%. Only",c*100,"% of total prediction has error above 20%")
```

```python
plt.plot(act_cnt2, label = "Actual")
plt.plot(pre_cnt2, label = "Predicted")
plt.xlabel('combined value of registred and casual bike usage')
plt.ylabel('Cnt')
plt.title('Combined value Prediction from Registered andCasual bike usage')
plt.legend()
plt.show()
```
